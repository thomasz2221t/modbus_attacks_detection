```python
def get_package_attributes(packet): 
  try:
    id = packet.ip.id
    time = packet.tcp.time_relative
    protocol = packet.ip.proto
    source_mac = packet.eth.addr
    source_address = packet.ip.src
    source_port = packet.tcp.srcport
    destination_mac = packet.eth.dst
    destination_address = packet.ip.dst
    destination_port = packet.tcp.dstport
    packet_length = packet.length
    ip_header_checksum = packet.ip.checksum
    ip_total_length = packet.ip.len
    ip_header_length = packet.ip.hdr_len
    ip_flags_register = packet.ip.flags
    ip_ttl = packet.ip.ttl
    tcp_checksum = packet.tcp.checksum
    tcp_header_length = packet.tcp.hdr_len
    tcp_total_length = packet.tcp.len
    tcp_flags = packet.tcp.flags
    tcp_timestamp_echoreply = packet.tcp.options_timestamp_tsecr
    tcp_timestamp_value = packet.tcp.options_timestamp_tsval
    previous_frame_time = packet.tcp.time_delta
    frame_vector = [id, time, protocol, source_mac, source_address, source_port, destination_mac, destination_address, destination_port, packet_length, ip_header_checksum,
                    ip_total_length,ip_header_length, ip_flags_register, ip_ttl, tcp_checksum, tcp_header_length, tcp_total_length, tcp_flags, 
                    tcp_timestamp_echoreply, tcp_timestamp_value, previous_frame_time]
    return frame_vector
  except AttributeError as e:
    pass
```

```python
from datetime import datetime
from dateutil.parser import parse
def getDateParametersFromPacket(attack_time, send_time):
  send_time = send_time[:27].strip()
  time_send = datetime.strptime(send_time,"%b %d, %Y %H:%M:%S.%f")
  time_attack = datetime.strptime(attack_time, "%Y-%m-%d %H:%M:%S.%f")
  difference = time_send - time_attack
  if((abs(difference.total_seconds())/60) < 30):
    print(difference.total_seconds())
    print(abs(difference.total_seconds())/60)
    print('yes')
    return True
  else:
    return False
```

```python
def getDeviceDataset(dir, frame_id, attack_time):
  print('0x'+'{:04x}'.format(frame_id))
  frames = []
  print("Loading directory: ", dir)
  for file in os.listdir(dir):
    file_dir = dir + '\\' + file
    print(file_dir)
    data = pd.read_csv(file_dir, delimiter=';')
    array = np.array(data)
    i = 0
    for row in array: 
      i = i + 1
      if row[0] == '0x'+'{:04x}'.format(frame_id) and getDateParametersFromPacket(attack_time, row[1]):
        print(row)
        frames.append(array[i-1])
        frames.append(row)
        frames.append(array[i+1])
        frames.append(array[i+2])
  return frames
```
